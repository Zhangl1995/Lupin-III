## question3. HR diagram

```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import math

#read the photometry data of M53&M30 
n_53,V_m53,B_V_53 = np.loadtxt('c:/users/a/desktop/stellar-structure/M53/table2.dat',usecols=(0,3,4),unpack = True)
n_30,V_m30,B_V_30 = np.loadtxt('c:/users/a/desktop/stellar-structure/M30/table1.dat',usecols=(0,4,5),unpack = True)

#calculate the absolute V-band absolute maginitude
V_53 = V_m53-5*(math.log(17790,10)-1)
V_30 = V_m30-5*(math.log(8312,10)-1)

#plot the data
plt.subplot(1,1,1)

plt.plot(B_V_53,V_53,'r+',label='M53')
plt.plot(B_V_30,V_30,'b+',label='M30')

#add the star in 2 in the HR diagram
#the B-V color index and absolute magnitude is calculated with the help of information
#such as the star's exact star type(I think it means temperature or color index), V band magnitude and redshift from the fits head
#also Ballesteros' formula in Ballesteros, F. J. (2012) to convert T to B-V,
#and Wright (2006, PASP, 118, 1711) to convert z to DL to get absolute magnitude
plt.plot(0.2,-4.39,'y*')
plt.text(0.2,-4.39,'star in 2',color='y')
#Mark the location of 7 types of star according to color index(the number is searched from
#wikipedia that referring to Zombeck, Martin V. (1990). "Calibration of MK spectral types". Handbook of Space Astronomy and Astrophysics (2nd ed.)
plt.axvline(-0.33,color='k')
plt.axvline(-0.30,color='k')
plt.axvline(-0.02,color='k')
plt.axvline(0.30,color='k')
plt.axvline(0.58,color='k')
plt.axvline(0.81,color='k')
plt.axvline(1.40,color='k')
plt.text(-0.35,-3,'O5')
plt.text(-0.32,-2,'B5')
plt.text(-0.04,-3,'A5')
plt.text(0.28,-2,'F5')
plt.text(0.56,-3,'G5')
plt.text(0.79,-2,'K5')
plt.text(1.38,-3,'M5')

plt.xlim(-1.0,3.0)
plt.gca().invert_yaxis()
plt.xlabel('B-V color index',size=20)
plt.ylabel('V absolute magnitude',size=20)
plt.legend()

plt.show()
```
