## 3.Introduction to the Saha Equation

### (a)

name the $\frac{ionized\ H}{ionized\ H + neutral\ H}$ as  y

by Saha equation, the eqution bwlow can be get

$\frac{y^{2}}{1-y} = \frac{1}{n}(\frac{2\pi m_{e}kT}{h^{2}})^{}3/2 e^{-\chi_{H}/kT} = \frac{4\times10^{-9}g cm^{-3}}{\rho}T^{3/2}e^{-1.578\times10^{5}/T}$

to every given y, we can get the relation between T and $\rho$

equally, we can get the relation of y versus T by every given $\rho$

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(0.1,1,0.10)
logT = np.arange(3,7,0.1)

for i in range(0,len(y)):
    rho = 4*10**-9*(10**logT)**1.5*np.exp(-1.578*10**5/(10**logT))*(1-y[i])/y[i]**2
    plt.plot(np.log10(rho),logT,label='ioni. frac.='+np.str(y[i]))
plt.xlim(-9,0.3)
plt.xlabel('log(rho)',size=10)
plt.ylabel('logT',size=10)
plt.legend(loc=2,fontsize = 'small')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

y = Symbol('y')
#logrho = np.array([0.01])
logrho = np.hstack((np.arange(-7,0.3,1),0.3))
logT = np.arange(3.5,6.5,0.1)

for i in range(0,len(logrho)):
#solve the ioni. frac. by given rho and series of T
    fra = np.array([solve(4*10**-9*(10**each)**1.5*np.exp(-1.578*10**5/(10**each))/10**logrho[i] - y**2/(1-y),y) for each in logT])
#here onething need to notice is that the logrho can't be too small for this code(<10^-8)
    fra = [each[1] for each in fra]
    fra = np.array(fra)
#    print(fra)
    plt.plot(logT,fra,label='log(rho)'+ '='+ np.str(logrho[i]))
    
plt.xlabel('logT',size=10)
plt.ylabel('inoization fraction',size=10)
plt.legend(loc=4,fontsize = 'small')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii

dat = ascii.read('homework2csv/Beijing_2008.csv',delimiter=',')
value = dat['Value']
QC = dat['QC Name']
d = np.arange(0,len(dat['Value']),24)
data = [np.sum(value[d[i]:d[i+1]])/len(QC[d[i]:d[i+1]][QC[d[i]:d[i+1]]=='Valid']) for i in range(0,len(d)-1)]
data = np.hstack((data,np.sum(value[d[-1]:])/len(QC[d[-1]:][QC[d[-1]:]=='Valid'])))
#print(len(d))
#print(len(data))
plt.plot(d/24+99.6,data,label='2008')

y = np.arange(2009,2017,1)
for n in range(0,len(y)):
    dat = ascii.read('homework2csv/Beijing_'+ np.str(y[n])+'.csv',delimiter=',')
    value = dat['Value']
    QC = dat['QC Name']
    d = np.arange(0,len(dat['Value']),24)
    data = [np.sum(value[d[i]:d[i+1]])/len(QC[d[i]:d[i+1]][QC[d[i]:d[i+1]]=='Valid']) for i in range(0,len(d)-1)]
    data = np.hstack((data,np.sum(value[d[-1]:])/len(QC[d[-1]:][QC[d[-1]:]=='Valid'])))
#    print(len(d))
#    print(len(data))
    plt.plot(d/24,data,label=np.str(y[n]))
plt.ylabel('mean Value of PM2.5 (ug/cm^3)')
plt.xlabel('date(day)')
plt.legend(fontsize = 'small')
plt.xlim(0,450)
plt.ylim(0,)
plt.show()
```
