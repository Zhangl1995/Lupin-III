```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

gala_list = open('c:/users/a/desktop/galaxy/assignment3/galaxies2.txt','r')
gala_name = gala_list.readlines()
gala_list.close()
#read the data and plot
for i in range(0,len(gala_name)):
#    plt.figure(figsize=(12,6))
    sdu = fits.open('c:/users/a/desktop/galaxy/assignment3/'+gala_name[i].strip())
    plt.plot(10**sdu[1].data['loglam'],sdu[1].data['flux']*(10**-17),'k')
    plt.xlabel('$\lambda(\AA) $')
    plt.ylabel('Flux($erg/(cm^{2} sec Hz)$)')
    plt.show()
```
