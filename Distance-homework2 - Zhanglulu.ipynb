```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np

data = ascii.read('SeboLMC.dat')
mask = data['Period']>0
logP = np.log10(data['Period'][mask])
Mv = data['Vmag'][mask]

coe1 = np.polyfit(logP,Mv,1)
coe2 = np.polyfit(logP,Mv,2)
fit1 = np.poly1d(coe1)
fit2 = np.poly1d(coe2)

sigma_V = 0.2
Mv_1 = fit1(logP)
Mv_2 = fit2(logP)
C_S1 = np.sum(((Mv - Mv_1)/sigma_V)**2)
C_S2 = np.sum(((Mv - Mv_2)/sigma_V)**2)

print('values of c\n linear fit     c1:%-10.3f   c2:%-10.3f\n quadratic fit  c1:%-10.3f   c2:%-10.3f   c3:%-10.3f'%(coe1[0],coe1[1],coe2[0],coe2[1],coe2[2]))
print('chi-squre\n  linear fit：  %-10.3f  quadratic fit: %-10.3f'%(C_S1,C_S2))

plt.scatter(logP,Mv,s=2,c='black',label='samples')
plt.scatter(logP,Mv_1,marker='+',s=20,c='b',label='linear fit')
plt.scatter(logP,Mv_2,marker='x',s=20,c='c',label='quadratic fit')
plt.gca().invert_yaxis()
plt.legend(loc=2)
plt.xlim(-0.5,2.0)
plt.xlabel('logP (d)')
plt.ylabel('Vmag (mag)')
plt.show()
```

It shows that the quadratic fit gives a better fitting of the data , and I think it doesn't just means that the relation between Vmag and logP should be quadratic, but just means that the relation between Vmag data and logP data is not strictly linear and need some correction.

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np

#data = ascii.read('SeboLMC.dat')

#mask0 = data['Period']>0
#logP = np.log10(data['Period'][mask0])
#Mv = data['Vmag'][mask0]

data = ascii.read('SeboLMC.dat')
mask1 = data['Period']>0.5
mask2 = data['Vmag']<17
mask = mask1*mask2
logP = np.log10(data['Period'][mask])
Mv = data['Vmag'][mask]

mask = np.zeros(len(logP),dtype=bool)
for i in range(0,len(mask)):
    if(i%5 == 4):
        mask[i] = True

logP = logP[mask]
Mv = Mv[mask]

coe1 = np.polyfit(logP,Mv,1)
coe2 = np.polyfit(logP,Mv,2)
fit1 = np.poly1d(coe1)
fit2 = np.poly1d(coe2)

sigma_V = 0.2
Mv_1 = fit1(logP)
Mv_2 = fit2(logP)


#print('values of c\n linear fit     c1:%-10.3f   c2:%-10.3f\n quadratic fit: c1:%-10.3f   c2:%-10.3f   c3:%-10.3f'%(coe1[0],coe1[1],coe2[0],coe2[1],coe2[2]))

plt.scatter(logP,Mv-Mv_1,marker='+',s=20,c='b',label='linear fit')
plt.legend(loc=2)
plt.xlim(-0.5,2.0)
plt.xlabel('logP (d)')
plt.ylabel('$r_{i}$ (mag)')
plt.show()

plt.scatter(logP,Mv-Mv_2,marker='x',s=20,c='c',label='quadratic fit')
#plt.gca().invert_yaxis()
plt.legend(loc=2)
plt.xlim(-0.5,2.0)
plt.xlabel('logP (d)')
plt.ylabel('$r_{i}$ (mag)')
plt.show()

plt.scatter(logP,Mv-Mv_1,marker='+',s=20,c='b',label='linear fit')
plt.scatter(logP,Mv-Mv_2,marker='x',s=20,c='c',label='quadratic fit')
plt.legend(loc=2)
plt.xlim(-0.5,2.0)
plt.xlabel('logP (d)')
plt.ylabel('$r_{i}$ (mag)')
plt.show()
```

When I compare my calculated value with the current consensus value in the literature, I find it's much larger than it. With a explore of the scatter plotting of the data, I think the difference may be result from the bad samples, such as those with low magnitude, so I restrict the Vmag data with its number value small than 17 and calculated again, then I get a much better result as shown in below($\chi^{2}$ decrease from ~7000 to ~2000). For there is still some difference between the calculated value with the current consensus value, I think more detailed correction should be done.

```python
from astropy.io import ascii
import matplotlib.pyplot as plt
import numpy as np

data = ascii.read('SeboLMC.dat')
mask1 = data['Period']>0.5
mask2 = data['Vmag']<17
mask = mask1*mask2
logP = np.log10(data['Period'][mask])
Mv = data['Vmag'][mask]

coe1 = np.polyfit(logP,Mv,1)
coe2 = np.polyfit(logP,Mv,2)
fit1 = np.poly1d(coe1)
fit2 = np.poly1d(coe2)

sigma_V = 0.2
Mv_1 = fit1(logP)
Mv_2 = fit2(logP)
C_S1 = np.sum(((Mv - Mv_1)/sigma_V)**2)
C_S2 = np.sum(((Mv - Mv_2)/sigma_V)**2)

print('values of c\n linear fit     c1:%-10.3f   c2:%-10.3f\n quadratic fit  c1:%-10.3f   c2:%-10.3f   c3:%-10.3f'%(coe1[0],coe1[1],coe2[0],coe2[1],coe2[2]))
print('chi-squre\n  linear fit：  %-10.3f  quadratic fit: %-10.3f'%(C_S1,C_S2))

plt.scatter(logP,Mv,s=2,c='black',label='samples')
plt.scatter(logP,Mv_1,marker='+',s=20,c='b',label='linear fit')
plt.scatter(logP,Mv_2,marker='x',s=20,c='c',label='quadratic fit')
plt.gca().invert_yaxis()
plt.legend(loc=2)
plt.xlim(-0.3,)
plt.ylim(17.5,11)
plt.xlabel('logP (d)')
plt.ylabel('Vmag (mag)')
plt.show()
```
